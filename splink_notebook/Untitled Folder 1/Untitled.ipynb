```python
import findspark
findspark.init()

import pyspark
from pyspark.sql import SparkSession
from pyspark.context import SparkContext
from pyspark.sql.functions import *
from pyspark.sql.session import SparkSession
import re
```

```python
print("PySpark Application Started")

spark = SparkSession \
    .builder \
    .appName("Splink") \
    .master("local[*]") \
    .getOrCreate()

print(spark.sparkContext.appName)

df = spark.read.option("header", "true").csv("/home/abhishekks446/data/pre_test/mobile-update.csv")
# df = spark.read.option("header", "true").csv("/home/abhishekks446/data/pre_test/mobile-update.csv")
```

```python
schema_len=(len(df.schema.names))
```

```python
column_lst=[]
def no_special_characters(s, pat=re.compile('[@!#$%^&*()<>?/\|}{~:]')):
    if pat.search(s):
        column_lst.append(s)
for element in df.schema.names:
    no_special_characters(element)
print(column_lst)
```
