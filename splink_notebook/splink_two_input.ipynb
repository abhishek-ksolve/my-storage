```python
import findspark
findspark.init()

import pyspark
from pyspark.sql import SparkSession
from pyspark.context import SparkContext
from pyspark.sql.functions import *
from pyspark.sql.session import SparkSession
from splink import Splink
from splink.jar_fallback import jw_sim_py
```

```python
print("PySpark Application Started")

spark = SparkSession \
    .builder \
    .appName("Splink") \
    .master("local[*]") \
    .getOrCreate()


print(spark.sparkContext.appName)

df1 = spark.read.option("header", "true").csv("/home/abhishekks446/data/demo_20.csv").withColumn("source_dataset",lit("df_1"))
df2 = spark.read.option("header", "true").csv("/home/abhishekks446/data/demo_20.csv").withColumn("source_dataset",lit("df_2"))
```

```python
settings = {
    "link_type": "link_and_dedupe",
    "blocking_rules": [
        "l.venue = r.venue and l.year = r.year",
        "l.venue = r.venue and l.title = r.title"

    ],
    "comparison_columns": [

        {
            "col_name": "title",
            "num_levels": 2,
            "term_frequency_adjustments": True
        },
         {
            "col_name": "venue",
          
        }
       
    ]

}

linker = Splink(settings,[df1,df2], spark)
df_e = linker.get_scored_comparisons()
```

```python
df_new=df_e.toPandas().sort_values("match_probability", ascending=False)
```
